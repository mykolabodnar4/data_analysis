# Transfer learning

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from tensorflow.keras import layers
import os
```

### Getting the data

```python
tfds.disable_progress_bar()

((train_ds, validation_ds, test_ds), info) = tfds.load(
    "caltech101",
    # Reserve 10% for validation and 10% for test
    split=["train[:80%]", "train[80%:]", "test"],
    as_supervised=True,  # Include labels
    with_info=True
)

num_classes = 102

print(f"Number of training samples: {train_ds.cardinality()}")
print(f"Number of validation samples: {validation_ds.cardinality()}")
print(f"Number of test samples: {test_ds.cardinality()}")
```

```python
path = info.data_dir
classes = 'label.labels.txt'
clasess = os.path.join(path, classes)
text_file = open(clasess, "r")
class_names = text_file.readlines()
class_names = [x.strip() for x in class_names]
class_names
```

```python
train_ds
```

```python
plt.figure(figsize=(10, 10))
for i, (image, label) in enumerate(train_ds.take(9)):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title(class_names[int(label)])
    plt.axis("off")
```

### Standardizing the data

```python
num_labels = 102
resize_fn = keras.layers.Resizing(128, 128)

train_ds = train_ds.map(lambda x, y: (resize_fn(x), keras.utils.to_categorical(y, num_labels)))
validation_ds = validation_ds.map(lambda x, y: (resize_fn(x), keras.utils.to_categorical(y, num_labels)))
test_ds = test_ds.map(lambda x, y: (resize_fn(x), keras.utils.to_categorical(y, num_labels)))
```

### Data augmentation

```python
IMG_SIZE = (128,128,3)


augmentation_layers = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
])
```

```python
for image, l in train_ds.take(1):
    plt.figure(figsize=(10, 10))
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        augmented_image = augmentation_layers(tf.expand_dims(image, 0))
        plt.imshow(np.array(augmented_image[0]).astype("int32"))
        plt.axis("off")
```

```python
from tensorflow import data as tf_data

batch_size = 64

train_ds = train_ds.batch(batch_size).prefetch(tf_data.AUTOTUNE).cache()
validation_ds = validation_ds.batch(batch_size).prefetch(tf_data.AUTOTUNE).cache()
test_ds = test_ds.batch(batch_size).prefetch(tf_data.AUTOTUNE).cache()
```

### Load MobileV2

```python
base_model = keras.applications.MobileNetV2(
    weights="imagenet",  # Load weights pre-trained on ImageNet.
    input_shape=IMG_SIZE,
    include_top=False,
)

# Freeze the base_model
base_model.trainable = False
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input # to rescale from (0, 255) to a range of (-1., +1.)

prediction_layer = tf.keras.layers.Dense(num_classes, activation='softmax')
inputs = keras.Input(shape=IMG_SIZE)

x = augmentation_layers(inputs)
x = preprocess_input(x)
x = base_model(x, training = False)
x = global_average_layer(x)
x = keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)

model = keras.Model(inputs, outputs)

model.summary(show_trainable=True)
```

```python
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
              loss=keras.losses.CategoricalCrossentropy(),
              metrics=["accuracy"])
```

```python
initial_epochs = 15

loss0, accuracy0 = model.evaluate(x = validation_ds)
```

```python
print("initial loss: {:.2f}".format(loss0))
print("initial accuracy: {:.2f}".format(accuracy0))
```

```python
history = model.fit(train_ds,
                    epochs=initial_epochs,
                    validation_data=validation_ds)
```

### Learning curves

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,3.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

## Fine tuning

```python
base_model.trainable = True
```

```python
print("Number of layers in the base model: ", len(base_model.layers))

# Fine-tune from this layer onwards
fine_tune_at = 100

# Freeze all the layers before the `fine_tune_at` layer
for layer in base_model.layers[:fine_tune_at]:
  layer.trainable = False
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate/10),
              metrics=[tf.keras.metrics.BinaryAccuracy(threshold=0.5, name='accuracy')])
```

```python
model.summary()
```

```python
fine_tune_epochs = 10
total_epochs =  initial_epochs + fine_tune_epochs

history_fine = model.fit(train_ds,
                         epochs=total_epochs,
                         initial_epoch=len(history.epoch),
                         validation_data=validation_ds)
```

```python
acc += history_fine.history['accuracy']
val_acc += history_fine.history['val_accuracy']

loss += history_fine.history['loss']
val_loss += history_fine.history['val_loss']
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.ylim([0.0, 1])
plt.plot([initial_epochs-1,initial_epochs-1],
          plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.ylim([0, 3.0])
plt.plot([initial_epochs-1,initial_epochs-1],
         plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

### Evaluation and prediction

```python
loss, accuracy = model.evaluate(test_ds)
print('Test accuracy :', accuracy)
```
